```python
from datetime import datetime

import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns
import json
```

```python
js_path = '../../datasets/bitly_usagov/example.txt'
records = (json.loads(r) for r in open(js_path, 'r'))
df = pd.DataFrame(records).replace('', np.NAN)
df.head()
```

```python
df['tz'].value_counts()
```

```python
tz = df['tz'].fillna('None').value_counts()
tz[-5:].append(tz[:5]).plot.bar()
```

```python
sns.barplot(y=tz[:5].index, x=tz[:5].values)
```

```python
df['os'] = np.where(df['a'].str.contains('Windows'),'Windows', 'Not Windows')
df.head()
```

```python
data = df.loc[:,['tz', 'os']]
data.head()
```

```python
gr = df.groupby(['tz', 'os'])
gr.tail()
```

```python
agg_counts = gr.size().unstack().fillna(0)
agg_counts.head()
```

```python
count_subset = agg_counts.sum(1).nlargest(n=10)
count_subset.name = 'total'
count_subset
```

```python
agg_counts
```

```python
count_subset
```

```python
count_subset.index.values
```

```python
res = pd.concat([agg_counts, count_subset],  axis=1).dropna().reset_index().sort_values('total')
res
```

```python
res.plot.bar()
```

```python
sns.barplot(x='total', y='tz',data=res)
```
